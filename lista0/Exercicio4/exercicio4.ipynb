```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from random import random as rnd

m = 10000                           # numero de elementos a ser criados
x = np.random.uniform(size=m)       # vetor coluna com elementos de dist aleatria uniforme
y = np.random.uniform(size=m)       # vetor coluna com elementos de dist aleatria uniforme

z = x + y

fig, ax = plt.subplots()

ax.hist(z, bins=100, density=1, color='royalblue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Histograma')

fig.tight_layout()
plt.show()
```
