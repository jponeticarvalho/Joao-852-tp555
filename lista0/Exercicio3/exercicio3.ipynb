```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from random import random as rnd

m = 1000                           # numero de elementos a ser criados
x = np.random.uniform(size=m)      # vetor coluna com elementos de dist aleatória uniforme
w = np.random.normal(0,1,m)        # Vetor coluna com elementos de dist. normal

y = 1.2 + 2.3*x + 10*w

fig, ax = plt.subplots()
ax.plot(x, y, '.')

ax.set(xlabel='input (x)', ylabel='output (y)',
       title='Gráfico 2D input e output')
ax.grid()

fig.savefig("grafico.png")
plt.show()
```
